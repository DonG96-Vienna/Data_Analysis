```python
df["Percentage_of_death"] = (df.Cumulative_deaths/df.Cumulative_cases)*100
```

```python
today = date.today()
yesterday = (today - timedelta(days = 1)).strftime("%Y-%m-%d") 
# Nur seit gestern anzeigen
# Gibt sicher bessere Methoden^^
```

```python
df.Percentage_of_death = df.Percentage_of_death.fillna(0)
# NA mit 0 füllen. Auch nicht ideal wahrscheinlich
```

```python
df.Percentage_of_death = pd.Series([round(val,2) for val in df.Percentage_of_death], 
                                   index = df.index)
# Runden auf 2 Nachkommastellen
```

```python
df.loc[df.Date_reported >= yesterday, ["Country", "Percentage_of_death"]]
```
